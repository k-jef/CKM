# Вариант 2. 
Численное вычисление интеграла
$$
\pi = \int_0^1\frac{4}{1+x^2}dx
$$
Нарисуем график этой функции на промежутке [0,1] 


```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from scipy import integrate
```

```python
xx = np.linspace(0, 1, 500)
yy = 4/(1+xx**2)
plt.plot(xx, yy)
plt.grid()
pass
```

1. Посчитаем интеграл с помощью формулы прямоугольников с 2,4,8,16,32,64,128 узлами.

```python
a = 0
b = 1
for n in [2,4,8,16,32,64,128]:
    h = (b - a)/n
    x = np.arange(a + h/2, b, h)
    y =4/(1+x**2)
    I_rect =h*sum(y)
    plt.plot(n,abs(np.pi-I_rect)/(np.pi),'o',color='tab:orange')
    print('Кол-во узлов ',n,' Погрешность ',abs(np.pi-I_rect)/(np.pi))
Ibest_rect=I_rect
Rectbest_err= abs(np.pi-I_rect)/(np.pi)
plt.yscale('log')
plt.xscale('log')
pass
Ibest_rect
```

2. Посчитаем интеграл с помощью формулы трапеции с 2,4,8,16,32,64,128 узлами.

```python
for n in [2,4,8,16,32,64,128]:
    h = (b - a)/n
    x = np.linspace(a, b, n + 1)
    y = 4/(1+x**2)
    I_trapz = integrate.trapz(y, x)
    plt.plot(n,(np.pi-I_trapz)/(np.pi),'o',color='tab:blue')
    print('Кол-во узлов ',n,' Погрешность ',(np.pi-I_trapz)/(np.pi))
plt.yscale('log')
plt.xscale('log')
pass
Trapzbest_err= abs(np.pi-I_trapz)/(np.pi)
Ibest_trapz=I_trapz

```

3. Посчитаем интеграл с помощью правила Симпсона с 2,4,8,16,32,64,128 узлами.

```python
for n in [2,4,8,16,32,64,128]:
    h = (b - a)/n
    x = np.linspace(a, b, n + 1)
    y = 4/(1+x**2)
    I_simp=integrate.simps(y)*h
    plt.plot(n,abs(np.pi-I_simp)/(np.pi),'o',color='tab:red')
    print('Кол-во узлов ',n,' Погрешность ',abs(np.pi-I_simp)/(np.pi))
plt.yscale('log')
plt.xscale('log')
pass
Simpbest_err= abs(np.pi-I_simp)/(np.pi)
Ibest_simp=I_simp
Ibest_simp
```

```python
print('Rectangaular method ', Ibest_rect,' ',Rectbest_err)
print('Trapezoidal method ', Ibest_trapz,' ',Trapzbest_err)
print('Simpson rule ', Ibest_simp,' ',Simpbest_err)
print('Pi', np.pi)

```

Подведем промежуточный итог:
1.Самым точным ,как и предсказывалось, стал интеграл посчитанный методом Симпсона(так как он имеет 4 порядок сходимости).
2.Стало достаточно неожидано, что интеграл, посчитанный с помощью правилом прямоугольников,стал более точным чем интеграл, посчитанный правилом трапеций. Изначально предсказывалась обратная ситуация, так как 1 способ имеет более маленький порядок сходимости.

# Вариант 3. 
Численное вычисление интеграла
$$
-\frac{4}{9} = \int_0^1 \sqrt{x}\ln x dx
$$
Нарисуем график этой функции на промежутке [0,1] 

```python
xx = np.linspace(1e-9, 1, 500)
yy = np.sqrt(xx)*np.log(xx)
plt.plot(xx, yy)
plt.grid()
pass
```

1. Посчитаем интеграл с помощью формулы прямоугольников с 2,4,8,16,32,64,128 узлами.

```python
a = 1e-12
b = 1
ans=-(4/9)
for n in [2,4,8,16,32,64,128]:
    h = (b - a)/n
    x = np.arange(a + h/2, b, h)
    y =np.sqrt(x)*np.log(x)
    I_rect =h*sum(y)
    plt.plot(n,(ans-I_rect)/abs(ans),'o',color='tab:orange')
    print('Кол-во узлов ',n,' Погрешность ',(ans-I_rect)/abs(ans))
Ibest_rect=I_rect
Rectbest_err= (ans-I_rect)/abs(ans)
plt.yscale('log')
plt.xscale('log')
pass
Ibest_rect
```

2. Посчитаем интеграл с помощью формулы трапеции с 2,4,8,16,32,64,128 узлами.

```python
for n in [2,4,8,16,32,64,128]:
    x = np.linspace(a, b, n + 1)
    y = np.sqrt(x)*np.log(x)
    I_trapz = integrate.trapz(y, x)
    plt.plot(n,(ans-I_trapz)/(ans),'o',color='tab:blue')
    print('Кол-во узлов ',n,' Погрешность ',(ans-I_trapz)/(ans))
plt.yscale('log')
plt.xscale('log')
pass
Trapzbest_err= (ans-I_trapz)/(ans)
Ibest_trapz=I_trapz
Ibest_trapz
```

3. Посчитаем интеграл с помощью правила Симпсона с 2,4,8,16,32,64,128 узлами.

```python
for n in [2,4,8,16,32,64,128]:
    h = (b - a)/n
    x = np.linspace(a, b, n + 1)
    y = np.sqrt(x)*np.log(x)
    I_simp=integrate.simps(y)*h
    plt.plot(n,(ans-I_simp)/(ans),'o',color='tab:red')
    print('Кол-во узлов ',n,' Погрешность ', abs((ans-I_simp)/(ans)))
plt.yscale('log')
plt.xscale('log')
pass
Simpbest_err= abs((ans-I_simp)/(ans))
Ibest_simp=I_simp
Ibest_simp
```

```python
print('Rectangaular method ', Ibest_rect,' ',Rectbest_err)
print('Trapezoidal method ', Ibest_trapz,' ',Trapzbest_err)
print('Simpson rule ', Ibest_simp,' ',Simpbest_err)
print('answer ',ans)
```

Подведем промежуточный итог:

1.Самым точным ,неожиданно, стал интеграл ,посчитанный с помощью правила прямоугольников.Так могло получилось из-за достаточно небольшого числа элементарных отрезков.


# Вывод

Таким образом видно , что порешности изменяются как $$f(x)=\frac{1}{x^n}$$
Только в 1 и 2 способе n меньше , чем в методе Симпсона.

Но почему же все таки метод Сипсона проиграл во втором случае? Для этого стоит обратить внимание на вид остаточных членов:

1.В интегале по формуле прямоугольников справедливо 
             $$ R = \frac{b-a}{24} f''(\xi)\, h^2, \quad \mbox{где $\xi \in [a, b]$}$$

2.В интеграле, почитанным методом Симпсона 
        $$
  R = \frac{b-a}{180} f^{IV}(\xi)\, h^4, \quad \mbox{где $\xi \in [a, b]$}.
$$
И видно, что такая ситуация могла произойти, когда  $$f^{IV} > f'' $$

Возьмем 2 и 4 производную и сравним их: 
$$
f^{IV}=\frac{16-15*ln(x)}{16*x^{7/2}}
$$

$$
f''= -\frac{ln(x)}{4*x^{3/2}}
$$
Нарисуем график$$ f^{IV}-f''$$



```python
xx = np.linspace(1e-7, 1, 500)
yy = (16-15*np.log(xx))/(16*xx**(7/2)) + (np.log(xx))/(4*xx**(3/2))
    
plt.plot(xx, yy)
plt.grid()
pass
```

Таким образом , видно , что четвертая производная сильно больше чем вторая производная . И из-за этого такая погрешность вычислений у метода Симпсона.

Так же видно , что с какого-то момента погрешности почти не изменяются. Это связанно, с тем что количество элементарных отрезков становится достаточно большим и точность измерений становится высокой. Так же это связанно с погрешностью вычислений чисел с плавующей запятой, которыми и являются наши переменные в программе.
